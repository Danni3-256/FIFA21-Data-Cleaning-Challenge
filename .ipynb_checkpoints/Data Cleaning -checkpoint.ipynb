# The Contract Column

```python
fifa21_V2['Contract'].head(2)
```

```python
fifa21_V2['Contract'].unique()
```

```python
import re

def contract_start_year(contract):
    if contract == "Free":
        return "Free"
    elif "On Loan" in contract:
        return "On Loan"
    else:
        match = re.search('(\d{4})\s*', contract)
        if match:
            return match.group()
        else:
            return None

fifa21_V2['Contract Start Year'] = fifa21_V2['Contract'].apply(contract_start_year)
fifa21_V2.head()
```

```python
import re

def contract_end_year(contract):
    if contract == "Free":
        return "Free"
    elif "On Loan" in contract:
        return "On Loan"
    else:
        match = re.search('\d{4}$', contract)
        if match:
            return match.group()
        else:
            return None

fifa21_V2['Contract End Year'] = fifa21_V2['Contract'].apply(contract_end_year)
fifa21_V2.head()
```

# The height column

```python
fifa21_V2['Height'].info()
```

```python
fifa21_V2['Height'].sample(20)
```

```python
fifa21_V2['Height'].unique() #We can see that we have height in cms and in feet and inches
```

```python
# #We need to have one uniform unit of measurement so let's convert the feet and inches to cms
# #1 foot = 30.48cms and 1 inch = 2.54cms
# Defining a function to convert heights to centimeters
def height_to_cm(height):
    if '\'' in height:
        feet, inches = height.split('\'')
        cm = round(int(feet) * 30.48 + int(inches.strip('"')) * 2.54,2) # Convert height from feet and inches to centimeters

    else:
        cm = int(height.strip('cm'))
    return cm

# Apply the conversion function to the 'Height' column
fifa21_V2['Height'] = fifa21_V2['Height'].apply(height_to_cm)
fifa21_V2['Height'].unique()
```

```python
fifa21_V2 = fifa21_V2.rename(columns={'Height': 'Height (Kgs)'})
```

# The Weight Column

```python
fifa21_V2['Weight'].sample(20)
```

```python
fifa21_V2['Weight'].unique()
```

```python
# Convert weight to kg
def weight_to_kg(weight):
    if 'kg' in weight:
        kg = float(weight.strip('kg'))
    elif 'lbs' in weight:
        lbs = float(weight.strip('lbs'))
        kg = round(lbs / 2.2046,2)
    else:
        kg = float('nan')
    return kg

fifa21_V2['Weight'] = fifa21_V2['Weight'].apply(weight_to_kg)
fifa21_V2['Weight']
```

# Value, Wage, Hits and Release Clause columns

```python
fifa21_V2.head(2)
```

<b>Value Column</b>

```python
fifa21_V2['Value'].unique()
```

All the values are in Euros. Going to change the column name to "Value (Euros)" and remove unwanted characters.  

```python
def value_to_numeric(value):
    multiplier = 1
    if value.endswith('M'):
        multiplier = 1000000
    elif value.endswith('K'):
        multiplier = 1000
    value = value.replace('€', '').replace('M', '').replace('K', '')
    value = float(value) * multiplier
    return value

fifa21_V2['Value'] = fifa21_V2['Value'].apply(value_to_numeric)
fifa21_V2['Value'].head()
```

```python
fifa21_V2 = fifa21_V2.rename(columns={'Value': 'Value (Euros)'})
```

<b>Wage Column</b>

```python
fifa21_V2['Wage'].unique()
```

```python
def wage_to_numeric(wage):
    multiplier = 1
    if wage.endswith('M'):
        multiplier = 1000000
    elif wage.endswith('K'):
        multiplier = 1000
    wage = wage.replace('€', '').replace('M', '').replace('K', '')
    wage = float(wage) * multiplier
    return wage

fifa21_V2['Wage'] = fifa21_V2['Wage'].apply(wage_to_numeric)
fifa21_V2['Wage'].head()
```

```python
fifa21_V2 = fifa21_V2.rename(columns={'Wage': 'Wage (Euros)'})
```

<b>Release Clause Column</b>

```python
fifa21_V2['Release Clause'].unique()
```

```python
def release_clause_to_numeric(release_clause):
    multiplier = 1
    if release_clause.endswith('M'):
        multiplier = 1000000
    elif release_clause.endswith('K'):
        multiplier = 1000
    release_clause = release_clause.replace('€', '').replace('M', '').replace('K', '').replace('B','')
    release_clause = float(release_clause) * multiplier
    return release_clause

fifa21_V2['Release Clause'] = fifa21_V2['Release Clause'].apply(release_clause_to_numeric)
fifa21_V2['Release Clause'].head()
```

```python
fifa21_V2 = fifa21_V2.rename(columns={'Release Clause': 'Release Clause (Euros)'})
fifa21_V2.head()
```

<b>Hits Column</b>

```python
fifa21_V2['Hits'].unique()
```

```python
fifa21_V2['Hits'].isna().sum()
```

```python
fifa21_V2['Hits'].info()
```

```python
def hits_to_numeric(hits):
    multiplier = 1
    hits = str(hits)
    if pd.isna(hits):
        return 0
    elif hits.endswith('K'):
        multiplier = 1000
    hits = hits.replace('K', '')
    hits = round(float(hits) * multiplier,2)
    return hits

fifa21_V2['Hits'] = fifa21_V2['Hits'].apply(hits_to_numeric)
fifa21_V2['Hits'].head()
```

```python
fifa21_V2['Hits'].isna().sum()
```

```python
fifa21_V2['Hits'].unique()
```

```python
# Replacing NaN values in the Hits column with the mean value
mean_hits = fifa21_V2['Hits'].mean()
fifa21_V2['Hits'].fillna(mean_hits, inplace=True)
```
